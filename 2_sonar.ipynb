```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data'
df = pd.read_csv(url, header=None)
df
```

```python
ds = df.values

X = ds[:, :-1].astype(float)
Y = ds[:, -1]
```

```python
encoder = LabelEncoder()
encoded_Y = encoder.fit_transform(Y)
```

```python
model = Sequential()
model.add(Dense(60, input_dim=60, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])

hist = model.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=60, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])

hist2 = model.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
```

```python
model = Sequential()
model.add(Dense(60, input_dim=60, kernel_initializer='normal', activation='relu'))
model.add(Dense(15, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])

hist3 = model.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
```

```python
model = Sequential()
model.add(Dense(500, input_dim=60, kernel_initializer='normal', activation='relu'))
model.add(Dense(15, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])

hist4 = model.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(223)
plt.plot(hist.history['val_accuracy'])
plt.plot(hist2.history['val_accuracy'])
plt.plot(hist3.history['val_accuracy'])
plt.legend(['(60, 1)', '(30, 1)', '(60, 15, 1)'])

plt.subplot(224)
plt.plot(hist.history['val_loss'])
plt.plot(hist2.history['val_loss'])
plt.plot(hist3.history['val_loss'])
plt.legend(['(60, 1)', '(30, 1)', '(60, 15, 1)'])

plt.tight_layout()
```
